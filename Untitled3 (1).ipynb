```python
df.drop('Unnamed: 0',axis=1,inplace=True)
df.drop('favoriteCount',axis=1,inplace=True)
```

```python
df['publishedAt'] = pd.to_datetime(df['publishedAt'])
df['Year'] = df['publishedAt'].dt.year
df['Month'] = df['publishedAt'].dt.month_name()
df['Day'] = pd.to_datetime(df['publishedAt']).dt.day_name()
```

```python
import re
```

```python
def duration_to_min(a):
    minutes_match=re.search(r'(\d+)M',a)
    second_match= re.search(r'(\d+)S',a)

    minutes= int(minutes_match.group(1)) if minutes_match else 0
    seconds= int(second_match.group(1)) if second_match else 0

    total_min= minutes+ seconds/60.0
    return total_min


df['duration']= df['duration'].apply(duration_to_min)
```
